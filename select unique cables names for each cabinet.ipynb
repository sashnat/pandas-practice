```python

import pandas as pd
from openpyxl import load_workbook
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import scipy.stats as stats
```

```python
way_data = r'C:\\Users\\database\\UTNISO_DB_21_03_19.xlsx'
```

```python
df = pd.read_excel(way_data, sheet_name='БД')
#df_w_na = df.dropna(subset=['FTA cable'])
field_fta_cab = ['FТA cabinet', 'FTA cable']
field_sys_cab = ['SYS cabinet', 'SYS cable']
df_fta = df[field_fta_cab]
df_sys = df[field_sys_cab]
df_fta.head(20)
```

```python
df_fta=df_fta.drop_duplicates(keep="last")
df_fta.head(20)
```

```python
res = r'C:\\Users\database\\res_21_03_19.xlsx'
```

```python
df_fta.to_excel(res, sheet_name='FTA cab')
```

```python
# to add one more sheet with new data into the existing file
df_sys=df_sys.drop_duplicates(keep="last")
book = load_workbook(res)
writer = pd.ExcelWriter(res, engine='openpyxl')
writer.book = book # < panda will just use the writer to write the data to a new book thus overwriting the existing book if you don't specify that the book already has data
df_sys.to_excel(writer, sheet_name='SYS cab')
writer.save()
```
