```python
import pandas as pd
```

```python
set1 = [10, 11, 12]
set2 = [25, 27, 30]
set3 = [12, 14, 15]

df1 = pd.DataFrame({'set1':set1, 'set2':set2, 'set3':set3})
df1
```

```python
all_set = [10, 11, 16, 25, 22, 30, 15]
set_codes = ['A', 'B', 'C', 'D', 'E', 'F', 'G']

df2 = pd.DataFrame({'all_set':all_set, 'set_codes':set_codes})
df2
```

```python
# make a dictionary to set 'keys-values" matching 
rename_dict = df2.set_index('all_set').to_dict()['set_codes']
rename_dict
```

```python
# make a new column using a key with different header names
df1['set1_code'] = df1['set1'].replace(rename_dict)
df1
```
