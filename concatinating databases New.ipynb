#The 1st variant

```python
df1 = pd.concat(list(pd.read_excel(table, sheet_name = "Sheet{}".format(i)) for i in range(1,5)))
```

```python
# instead of df1 = df1.fillna({'q-ty1':0,'q-ty2':0,'q-ty3':0,'q-ty4':0})

l_q = ['q-ty1','q-ty2','q-ty3','q-ty4']
d = {i:0 for i in l_q}
df1 = df1.fillna(d)
```

```python
# instead of "df1['sum_m']= df1['q-ty1'].astype(int) + df1['q-ty2'].astype(int) + df1['q-ty3'].astype(int) + df1['q-ty4'].astype(int)"

df1['sum_m']= sum(list(df1[i] for i in l_q))
```

```python
f = {'model': 'first', 'name' : 'first', 'code' : 'first', 
     'customer' : 'first', 'prod' : 'first', 'item' : 'first', 'sum_m': 'sum', 
     'sheet1' : 'first','sheet2' : 'first', 'sheet3' : 'first', 'sheet4' : 'first', }

df1['-'] = ['$' for i in range(len(df1))]
df1['name_model'] = df1[['name', '-', 'model']].astype(str).sum(axis=1)
df1 = df1.groupby('name_model', as_index=False).agg(f)

df1
```

# The variants placed below just to see how the code was developed

The 2d variant

```python
list_dfn = []
for i in range(1,5):

    dfn = pd.read_excel(table, sheet_name="Sheet{}".format(i))
    list_dfn.append(dfn)
```

```python

df2 = pd.concat(list_dfn)
df2['-'] = ['$' for i in range(len(df2))]
df2['name_model'] = df2[['name', '-', 'model']].astype(str).sum(axis=1)
df2 = df2.fillna({'q-ty1':0,'q-ty2':0,'q-ty3':0,'q-ty4':0})
df2['sum_m']= df2['q-ty1'].astype(int) + df2['q-ty2'].astype(int) + df2['q-ty3'].astype(int) + df2['q-ty4'].astype(int)

f = {'sum_m': 'sum', 'model': 'first', 'name' : 'first', 'code' : 'first', 'customer' : 'first', 'item' : 'first', 'prod' : 'first'}
df2 = df2.groupby('name_model', as_index=False).agg(f)
df2
```

The 3d variant

```python
df_1['-'] = ['$' for i in range(len(df_1))]
df_1['name_model'] = df_1[['name', '-', 'model']].astype(str).sum(axis=1)
df_2['-'] = ['$' for i in range(len(df_2))]
df_2['name_model'] = df_2[['name', '-', 'model']].astype(str).sum(axis=1)
df_3['-'] = ['$' for i in range(len(df_3))]
df_3['name_model'] = df_3[['name', '-', 'model']].astype(str).sum(axis=1)
df_4['-'] = ['$' for i in range(len(df_4))]
df_4['name_model'] = df_4[['name', '-', 'model']].astype(str).sum(axis=1)
```

```python
df1_2 = df_1.set_index('name_model').combine_first(df_2.set_index('name_model')).reset_index()
df1_3 = df1_2.set_index('name_model').combine_first(df_3.set_index('name_model')).reset_index()
df3= df1_3.set_index('name_model').combine_first(df_4.set_index('name_model')).reset_index()
df3
```

```python
df3['name'] = df3['name_model'].str.split('$').str[0]
df3 = df3.drop(['name_model','-'], axis=1)


```

```python
df3 = df3.fillna({'q-ty1':0,'q-ty2':0,'q-ty3':0,'q-ty4':0})
```

```python
df3['sum']= df3['q-ty1'].astype(int) + df3['q-ty2'].astype(int) + df3['q-ty3'].astype(int) + df3['q-ty4'].astype(int)
df3
```

The 4th variant
