# The task is finding all duplicate rows in a pandas dataframe

```python
# original data
import pandas as pd
import numpy as np
import collections
np.random.seed(0)

df = pd.DataFrame(np.random.randn(9,3), 
                  index=['one', 'two', 'three', 'four', 'five', 'sex', 'seven', 'eight', 'nine'],
                  columns=list('abc'))
```

```python
def df_X(x):
    df[x] = [0 if i % 2 == 0 else 7 for i in range(0, len(df['c']))]
    return df[x]
df_X('new')
#df = df.replace('N/A',np.nan) #  replace all the N/A values with NaN in columns
df.head(10)
```

```python
df.iloc[0, 3] = 10
df.iloc[1, 3] = 20
df.iloc[6, 3] = 30
df.iloc[8, 3] = 20
df.head(10)
```

```python
df1 = df[df.duplicated('new',keep=False)] # or df1 = df[df.new.duplicated(keep=False)]
print (df1)
#a = df.groupby('new').apply(lambda x: list(x.index))
#print (a)
```

```python
# if you have not faced this task before, maybe you would solve the task this way like I did :).
 
l1 = [item for item, count in collections.Counter(df['new']).items() if count > 1]
list_df = [df.loc[df['new'] == i] for i in l1 if i in df['new'].values] # list of DataFrames
result = pd.concat(list_df)
print (result)
```
