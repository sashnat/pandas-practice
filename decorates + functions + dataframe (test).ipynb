```python
# pip install -U scipy
import pandas as pd
import numpy as np
np.random.seed(0)

df = pd.DataFrame(np.random.randn(9,3), 
                  index=['one', 'two', 'three', 'four', 'five', 'sex', 'seven', 'eight', 'nine'],
                  columns=list('abc'))
df.head(10)
```

```python
def dec(fn):
    def wrapped(*args): 
        # декоратор не должен работать только для функций, кол-во переменных которых строго соответствует кол-ву аргументов декоратора
        # в данном случае, под переменной понимают каждый элемент из списка [x,y]
        # поэтому в качестве аргумента декоратора указано *args
        print ('decorator runs')
        print ("fn is %s" % (fn.__name__))
        return fn(*args) + 1
    return wrapped
```

```python
@dec
def df_x(x, y):
    df[x] = [0 if i % 2 == 0 else 7 for i in range(0, len(df['c']))]
    df[y] = [100 if i % 2 == 0 else 7 for i in range(0, len(df['c']))]
    return df


print("df_x.__name__== " + df_x.__name__)
dec_res = df_x('new',"new1")
print(type(dec_res))
print("function result: decorator makes a new DataFrame as a result")
dec_res.head(10)

```
